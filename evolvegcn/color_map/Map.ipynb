```python
import networkx as nx
import matplotlib.pyplot as plt
import pickle
import scipy.sparse
import folium
import branca.colormap as cm
```

```python
G_file = open('street_nodes_dataframe.pickle', 'rb')
G = pickle.load(G_file)
```

```python
G
```

```python
vec_file = open('MAP_SOFT_PREDICTIONS.txt')
vetor = vec_file.readlines()
vec_file.close()
```

```python
vector = []
for i in range(len(vetor)):
    vector.append(float(vetor[i].rstrip('\n')))
```

```python
linear = cm.LinearColormap(["green",'blue', "red"], vmin=min(vector), vmax=max(vector))

linear
```

```python
m = folium.Map(location=[-23.5505, -46.6333], zoomstart=10)

folium.TileLayer("openstreetmap").add_to(m)

# folium.GeoJson(limites_latlong).add_to(m)

# Add points from gdf1 and gdf2 to the map
for idx, row in G.iterrows():
    coords = [(pt[1], pt[0]) for pt in row["geometry"].coords]
    folium.PolyLine(coords, color=linear(vector[idx]), popup=str(row["name"])).add_to(m)

# Display the map
m.save(f'color_map.html')
```
